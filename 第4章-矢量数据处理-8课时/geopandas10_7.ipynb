```python
import geopandas as gpd
import numpy as np
```

```python
path_kunming = 'E:\QGIS Data\kunming.shp'
```

```python
kunming_gpd = gpd.read_file(path_kunming)
kunming_gpd.head()
```

#### 矢量数据可视化

```python
kunming_gpd.plot(column='dt_name')
```

```python
id_chenggong = np.where(kunming_gpd['dt_name']=='chenggong')[0]
id_chenggong
chenggong_gpd = kunming_gpd.iloc[id_chenggong]
chenggong_gpd.plot(color='blue')
```

##### 数据写出

写出只含特定字段矢量

```python
path_kunming_out_1 = 'E:\QGIS Data\geopandas\sl_out_1.shp'
kunming_out_1 = kunming_gpd[['dt_name','geometry']]
kunming_out_1.head()
```

```python
kunming_out_1.to_file(path_kunming_out_1)

```

#### 写出只含特定要素矢量数据

```python
path_kunming_out_2 = 'E:\QGIS Data\geopandas\sl_out_2.shp'
kunming_out_2 = kunming_gpd.iloc[id_chenggong]
kunming_out_2.to_file(path_kunming_out_2)
```
